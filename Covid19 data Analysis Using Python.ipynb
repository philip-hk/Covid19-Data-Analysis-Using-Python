```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print('Modules are imported.')
```

```python
corona_dataset_csv = pd.read_csv("corona_Confirmed_dataset.csv")
corona_dataset_csv.head()
```

```python
corona_dataset_csv.shape
```

```python
df = corona_dataset_csv.drop(["Lat", "Long"], axis=1)
df.head()
```

```python
corona_dataset_csv.head()
```

```python
corona_dataset_csv.drop(["Lat", "Long"], axis=1, inplace = True)
corona_dataset_csv.head()
```

```python
# Aggregating the rows by the Country
corona_dataset_aggregated = corona_dataset_csv.groupby("Country/Region").sum()
corona_dataset_aggregated.head()
```

```python
corona_dataset_aggregated.shape
```

```python
# Visualizing data related to a country e.g. China
corona_dataset_aggregated.loc["China"]
```

```python
corona_dataset_aggregated.loc["China"].plot()
corona_dataset_aggregated.loc["Italy"].plot()
corona_dataset_aggregated.loc["Spain"].plot()
plt.legend()
```

```python
# Calculating a good Measure
# Good measure to describe the spread of the virus in a country
corona_dataset_aggregated.loc["China"].plot()
```

```python
# First 3 days infection cases
corona_dataset_aggregated.loc["China"][:3].plot()
```

```python
# Calculating and plotting the first derivative of the curve
corona_dataset_aggregated.loc["China"].diff().plot()
```

```python
# Maximum infection rate for China, Italy and Spain
corona_dataset_aggregated.loc["China"].diff().max()
```

```python
corona_dataset_aggregated.loc["Italy"].diff().max()
```

```python
corona_dataset_aggregated.loc["Spain"].diff().max()
```

```python
# Finding Maximum infection rate for all of the countries
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
max_infection_rates
```

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for c in countries:
    max_infection_rates.append(corona_dataset_aggregated.loc[c].diff().max())
corona_dataset_aggregated["max_infection_rate"] = max_infection_rates
corona_dataset_aggregated.head()
```

```python
# Create a new dataframe with selected colume
corona_data = pd.DataFrame(corona_dataset_aggregated["max_infection_rate"])
corona_data.head()
```

```python
# Importing worldwide_happiness_report dataset
happiness_report_csv = pd.read_csv("worldwide_happiness_report.csv")
happiness_report_csv.head()
```

```python
# Let's drop the useless columns
useless_cols = ["Overall rank", "Score", "Generosity", "Perceptions of corruption"]
happiness_report_csv.drop(useless_cols, axis=1, inplace = True)
happiness_report_csv.head()
```

```python
# Changing the indices of the dataframe 
happiness_report_csv.set_index("Country or region", inplace = True)
happiness_report_csv.head()
```

```python
# Joining two datasets 
```

```python
# Corona Dataset:
corona_data.head()
```

```python
corona_data.shape
```

```python
# World Happiness Report Dataset:
happiness_report_csv.head()
```

```python
happiness_report_csv.shape
```

```python
# Inner Join
data = corona_data.join(happiness_report_csv, how = "inner")
data.head()
```

```python
# Correlation matrix
data.corr()
```

```python
# Visualization of the results
data.head()
```

```python
# Plotting GDP vs maximum infection rate

sns.scatterplot(x = data["GDP per capita"], y = data["max_infection_rate"])
```

```python
sns.scatterplot(x = data["GDP per capita"], y = np.log(data["max_infection_rate"]))
```

```python
sns.regplot(x = data["GDP per capita"], y = np.log(data["max_infection_rate"]))
```

```python
# Plotting Healthy life expectancy vs maximum infection rate
sns.regplot(x = data["Healthy life expectancy"], y = np.log(data["max_infection_rate"]))
```
